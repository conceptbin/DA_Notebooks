# Correlation, visualisation
Using Pandas for data, Seaborn for visualisation.

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
# read in the data from a website in CSV format using Pandas
iris = pd.read_csv('http://h2o-public-test-data.s3.amazonaws.com/smalldata/iris/iris_wheader.csv')
```

```python
iris.info() # print out the columns and their types and number in series format
iris.head() # then look at the head of the data frame
```

```python
# How many categories are here?
iris['class'].unique()
```

```python
# describe() shows basic distribution of numerical data columns.
iris.describe()
```

## Scatterplots
Widths and lengths of petals and sepals, side-by-side.

Classes shown in different colours.

```python

fig, axs = plt.subplots(1, 2, figsize=(18, 6))

sns.scatterplot(data=iris, x='petal_len', y='petal_wid', hue='class', ax=axs[0])
sns.scatterplot(data=iris, x='sepal_len', y='sepal_wid', hue='class', ax=axs[1])
```

## All petal lenghts/widths
Scatterplot across all classes.

```python
#sns.scatterplot(iris['petal_len'], iris['petal_wid'])
sns.scatterplot(data=iris, x='petal_len', y='petal_wid')
```

## Pairplot

```python
sns.pairplot(iris)
```

## Heatmap
Showing egree of correlation between different categories

```python
plt.figure(figsize=(16,12))
sns.heatmap(iris.corr(),annot=True)
plt.plot()
```

```python
# Calculate the Pearson Correlation Coefficient between all features

corr_df = iris.corr(method="pearson") 
display(corr_df)

sns.heatmap(corr_df, cmap=sns.diverging_palette(220, 10, as_cmap=True),
            xticklabels=corr_df.columns.values,
            yticklabels=corr_df.columns.values)
```
